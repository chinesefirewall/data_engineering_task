```python
# required code to temper errors
from psycopg2.extensions import ISOLATION_LEVEL_AUTOCOMMIT 
pgconn.set_isolation_level(ISOLATION_LEVEL_AUTOCOMMIT) 
```

```python
# create POLICY data  table 

pgcursor.execute("""
CREATE TABLE IF NOT EXISTS policy_table
(
    policy_id              SERIAL
  , policy_type            VARCHAR(50) NOT NULL
  , policy_start_date      DATE
  , policy_end_date        DATE
  , premium_amount         MONEY
  , customer_id            BIGINT
);

""")
```

```python
# create customer data  table 

pgcursor.execute("""
CREATE TABLE IF NOT EXISTS customer_table
(
    customer_id             SERIAL
  , customer_name           VARCHAR(500)
  , customer_email          text not null unique 
  , customer_address        VARCHAR
  , age                     INT
  , gender                  VARCHAR
  , occupation              VARCHAR
   , address       VARCHAR
);

""")
```

```python
# create customer data  table 

pgcursor.execute("""
CREATE TABLE IF NOT EXISTS claim_table
(
    claim_id             SERIAL
  , claim_date           DATE
  , claim_amount         MONEY 
  , policy_id            BIGINT
);

""")
```

```python
# Sending the dataframe into the postgreSQL database

claim_data.to_sql(name='claims_table', con=pgconn, if_exists='replace')

customer_data.to_sql(name='customer_table', con=pgconn, if_exists='replace')

policy_data.to_sql(name='policy_table', con=pgconn, if_exists='replace')

print(' data succesfully pushed to DataBase')
```
